```python
# import necessary module packages
from bokeh.io import output_file, show
from bokeh.plotting import figure
import pandas as pd
```

```python
# reading our crash file
df = pd.read_csv("./data/transbase_collisions.csv")
# df.head()
```

```python
# get unique collision category
collision_type = df["type_of_collision"].unique()
collision_type.sort()
```

```python
years = df["accident_year"].unique()
years.sort()
years = years.astype(str).tolist()
```

```python
# groupby the dataset using accident_year and type_of_collision. This would return a dataset, convert the dataset into list
df2=df.groupby(["accident_year","type_of_collision"]).sum()[["number_killed","number_injured"]].reset_index()
# df2
```

```python
df2["number_killed"][df2["accident_year"]==2014].tolist()
```

```python
collisions = collision_type.tolist()
y2014 = df2["number_killed"][df2["accident_year"]==2014].tolist()
y2015 = df2["number_killed"][df2["accident_year"]==2015].tolist()
y2016 = df2["number_killed"][df2["accident_year"]==2016].tolist()
y2017 = df2["number_killed"][df2["accident_year"]==2017].tolist()
y2018 = df2["number_killed"][df2["accident_year"]==2018].tolist()
```

```python
output_file("bar_stacked.html")
data = {'collisions' : collisions,
        '2014'   : y2014,
        '2015'   : y2015,
        '2016'   : y2016,
        '2017'   : y2017,
        '2018'   : y2018}

colors = ["#FFC0CB", "#6A5ACD", "#DC143C","#FF8C00","#32CD32"]
```

```python
p = figure(x_range=collisions, plot_width=1000, plot_height=500, title="Fatality Counts by Collision Type",
           toolbar_location=None, tools="hover", tooltips="$name @collisions: @$name")
p.vbar_stack(years, x='collisions', width=0.7, color=colors, source=data, legend=[(str(x)+str(" ")) for x in years])

p.y_range.start = 0
p.x_range.range_padding = 0.0
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.legend.location = "top_left"
p.legend.orientation = "horizontal"

show(p)
```
